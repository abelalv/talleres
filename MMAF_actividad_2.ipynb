### 2.4.2 Jugando con el computador!


Se quiere calcular el volumen del material del tanque esférico con radio exterior $R$ y radio interior $r$. Para ello usaremos la fórmula del volumen de la esfera $(3)$, restando el volumen de la esfera interior al volumen de la esfera exterior:

$$
V_{tanque} = \frac{4}{3}\pi R^3 - \frac{4}{3}\pi r^3
$$


haciendo algunas manipulaciones algebraicas podemos simplificar la expresión anterior a:


$$
V_{tanque} = \frac{4}{3}\pi (R^3 - r^3)
$$

**Comprobemos numéricamente que la expresión correcta es diferente a $$\frac{4}{3}\pi (R - r)^3$$**
