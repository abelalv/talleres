### 8.4.1 Jugando con el computador!

Un sensor mide la actividad eléctrica del corazón en instantes cada vez más cercanos a un pico (es decir, la contracción máxima). El valor medido en el instante $t_n=\frac{1}{n}$ segundos después del pico se modela como:
$$S_n=S_0\cdot\sin(\frac{\alpha}{n})$$


$S_0$ es una constante positiva que representa la amplitud máxima de la señal en el pico del latido (por ejemplo, del orden de 0.5 a 2 mV).

$\alpha>0$ es una constante que controla qué tan rápido cambia la señal cerca del pico (puedes pensarla como un parámetro de “escala temporal” o de “fase”, sin necesidad de entrar en detalles biomédicos).


$n=1,2,3,…$ es el número de la medición: a medida que $n$ aumenta, estamos observando la señal en instantes cada vez más cercanos al instante exacto del pico cardíaco.

Recuerda: no necesitas saber medicina, solo interpretar una señal que se está midiendo en el tiempo.
